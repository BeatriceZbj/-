4. 缺失值的处理

**DataFrame**


**数据重塑(stack, unstack, pivot)**

stack: 将数据的列转为行

unstack: 将数据的行转为列

pivot:轴向旋转 

````python
 data = {'Id': [101, 101, 102, 102, 103, 103], '物品': ['小刀','笔','水','面包','笔','饼干'],'数量': [4,3,5,4,6,5]}
````

````python
df = pd.DataFrame(data)
df
````

````python
# 如何进行unstack，即将数据的行转为列
df.set_index(['Id','物品']) # 先设置一个双重索引
````

````python
df.set_index(['Id','物品']).unstack() # 行就转列啦
````

````python
df.set_index(['Id','物品']).unstack().stack() # 列就转行啦
````

````python
df.pivot('Id','物品','数量') # 直接行转列了更方便
# pivot(index=None, columns=None, values=None)
````

````python
# 传入标准函数聚合
def ptp(arr):
    return arr.max() - arr.min()
df.groupby(['物品']).agg([ptp]) # 注意这里有一个中括号
````

````python
df.groupby(['物品']).agg(lambda x:x.max()-x.min())
````

````python
df.groupby(['物品']).agg([np.mean, ptp, np.sum]) # 可以放入多个函数用逗号隔开中括号括起来
````

````python
# 也可以不同的列用不同的聚合函数
mapping = {'数量': np.mean, 'Id': np.sum}
df.groupby('物品').agg(mapping)
````

````python
from datetime import datetime, timedelta
now = datetime.now()  # 获取当前时间     2020-11-10 17:41:21.012941
new_time = now + timedelta(days=10)  # 给now加10天,days为天     2020-11-20 17:41:21.012941
new_time1 = now + timedelta(seconds=10)  # 给now加10秒,seconds为秒   2020-11-10 17:41:31.012941
````

````python
from pandas.tseries.offsets import DateOffset
ts = pd.Timestamp('2017-01-01 09:10:11')
ts + DateOffset(months=3)
# Timestamp('2017-04-01 09:10:11')
ts + DateOffset(hours=2)
# Timestamp('2017-01-01 11:10:11')
ts + DateOffset()
# Timestamp('2017-01-02 09:10:11')
````
